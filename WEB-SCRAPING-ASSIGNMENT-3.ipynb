### Que.1 Write a python program which searches all the product under a particular product vertical from www.amazon.in. The product verticals to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
#importing chromedriver
driver = webdriver.Chrome(r"C:\Users\ashis\Downloads\chromedriver_win32\chromedriver.exe")

url = 'https://www.amazon.in' #getting the required url
driver.get(url)

item = input("What you want to search ") #Input field

search_lap = driver.find_element_by_id("twotabsearchtextbox") #search box
search_lap.send_keys(item)

time.sleep(2) #wait for 2 m.s.
```

### Que.2 In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“

```python
search_btn = driver.find_element_by_xpath("//input[@id='nav-search-submit-button']") #submitting Laptop
search_btn.click()
time.sleep(2)

#collecting all the Product URLS
urls = []
for i in range(0,3):
    url_links=driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-4']/a")#collecting urls of all the laptop
    for i in url_links:
        urls.append(i.get_attribute('href'))
    nxt=driver.find_element_by_xpath("//li[@class='a-last']/a") #next button
    url=nxt.get_attribute('href')
    driver.get(url)
    time.sleep(2)
    
item_name = []
brand_name = []
Rating = []
No_Ratings = []
Price = []
Return = []
Expected_Delivery = []
Availability = []
Other_Details = []
for i in urls:
    driver.get(i)
    time.sleep(2)
    
    item=driver.find_element_by_xpath("//span[@id='productTitle']") #item name
    item_name.append(item.text)
    
    try:
        brand=driver.find_element_by_xpath("//div[@id='bylineInfo_feature_div']/div/a") #brand name
        brand_name.append(brand.text)
    except NoSuchElementException as e:
        brand_name.append("-")
    
    try:
        rate=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']") #Rating
        Rating.append(rate.text)
    except NoSuchElementException as e:
        Rating.append("-")
    
    try:
        no_rate=driver.find_element_by_xpath("//div[@id='averageCustomerReviews']/span[3]/a/span") #No of Ratings
        No_Ratings.append(no_rate.text)
    except NoSuchElementException as e:
        No_Ratings.append("-")
    
    try:
        MRP=driver.find_element_by_xpath("//span[@id='priceblock_ourprice']") #Price
        Price.append(MRP.text)
    except NoSuchElementException as e:
        try:
            MRP=driver.find_element_by_xpath("//span[@id='priceblock_dealprice']") #Price
            Price.append(MRP.text)
        except NoSuchElementException as e:
            Price.append("-")
    
    try:
        ret=driver.find_element_by_xpath("//div[@data-name='RETURNS_POLICY']/div[2]/a") #Return
        Return.append(ret.text)
    except NoSuchElementException as e:
        try:
            ret=driver.find_element_by_xpath("//div[@data-name='RETURNS_POLICY']/span/div[2]/a") #Return
            Return.append(ret.text)
        except NoSuchElementException as e:
            Return.append("-")
    
    try:
        Expected=driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b") #Expected_Delivery
        Expected_Delivery.append(Expected.text)
    except NoSuchElementException as e:
        Expected_Delivery.append("-")
    
    try:
        stock=driver.find_element_by_xpath("//div[@id='availability']/span") #Availability
        Availability.append(stock.text)
    except NoSuchElementException as e:
        Availability.append("-")
    
    try:
        features=driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']") #Other_Details
        Other_Details.append(features.text)
    except NoSuchElementException as e:
        Other_Details.append("-")
        
Oxygen=pd.DataFrame({}) #Dataframe
Oxygen['Name'] = item_name
Oxygen['Brand'] = brand_name
Oxygen['Rating'] = Rating
Oxygen['No of Ratings'] = No_Ratings
Oxygen['Price'] = Price
Oxygen['Return'] = Return
Oxygen['Expected_Delivery'] = Expected_Delivery
Oxygen['Availability'] = Availability
Oxygen['Other_Details'] = Other_Details
Oxygen['Urls'] = urls
```

### Que5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
import selenium
from selenium import webdriver
import re
import time

#connecting the web driver.
driver=webdriver.Chrome(r'C:\Users\ashis\Downloads\chromedriver_win32\chromedriver.exe')

#getting the required url
driver.get('https://www.google.com/maps')

#finding the search bar
search_bar = driver.find_element_by_xpath('//input[@class="tactile-searchbox-input"]')    
search_bar.clear()                                               
search_bar.send_keys(input('Enter the city name: ')) 

#finding the search button and clicking
search_button = driver.find_element_by_xpath('//button[@id="searchbox-searchbutton"]')       
search_button.click()
time.sleep(3)

#getting the current url
Location=driver.current_url

try:
    coordinates=re.findall(r'@(.*)data',Location)
    lat_lon=coordinates[0].split(",")
    lat=lat_lon[0]
    lon=lat_lon[1]
except IndexError as e:
    coordinates=re.findall(r'@(.*)z',Location)
    lat_lon=coordinates[0].split(",")
    lat=lat_lon[0]
    lon=lat_lon[1]
print("\nlatitude={},longitude={}".format(lat,lon))
```

```python
#creating empty list
Product=[]
Processor=[]
Memory=[]
OS=[]
Display=[]
Seller=[]
Price=[]
Weight=[]
Dimension=[]
Graphics_Processor=[]
```

```python
#finding the search button and clicking
laptop=driver.find_element_by_xpath("//div[@class='col-sm-6']/div[3]/a")
best_laptop=laptop.get_attribute('href')
driver.get(best_laptop)
```

```python
#scraping all the details
Brand_tag=driver.find_elements_by_xpath("//div[@class='right-container']/div/a/h3")
for i in Brand_tag:
    brandtag=i.text
    Product.append(brandtag)
operating = driver.find_elements_by_xpath("//div[@class='Product-specs']/div[2]/div/ul/li[1]/div/div")
for i in operating:
    oper = i.text
    OS.append(oper)
display = driver.find_elements_by_xpath("//div[@class='Product-specs']/div[2]/div/ul/li[2]/div/div")
for i in display:
    disp = i.text
    Display.append(disp)
processor = driver.find_elements_by_xpath("//div[@class='Product-specs']/div[2]/div/ul/li[3]/div/div")
for i in processor:
    proc = i.text
    Processor.append(proc)
memory = driver.find_elements_by_xpath("//div[@class='Product-specs']/div[2]/div/ul/li[4]/div/div")
for i in memory:
    mem = i.text
    Memory.append(mem)
weight = driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[7]/td[3]")
for i in weight:
    weig = i.text
    Weight.append(weig)
dimension = driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[8]/td[3]")
for i in dimension:
    dimen = i.text
    Dimension.append(dimen)
graphics = driver.find_elements_by_xpath("//div[@class='Spcs-details']/table/tbody/tr[9]/td[3]")
for i in graphics:
    grap = i.text
    Graphics_Processor.append(grap)
for i in range(2,9):
    sell = driver.find_element_by_xpath("//table[@id='summtable']/tbody/tr["+str(i)+"]/td[2]")
    Seller.append(sell.text)
    price = driver.find_element_by_xpath("//table[@id='summtable']/tbody/tr["+str(i)+"]/td[3]")
    Price.append(price.text)
```

```python
#creating the dataset
Top_10_Gaming_Laptop=pd.DataFrame({})
Top_10_Gaming_Laptop['Product']=Product
Top_10_Gaming_Laptop['Processor']=Processor
Top_10_Gaming_Laptop['Memory']=Memory
Top_10_Gaming_Laptop['OS']=OS
Top_10_Gaming_Laptop['Display']=Display
Top_10_Gaming_Laptop['Weight']=Weight
Top_10_Gaming_Laptop['Dimension']=Dimension
Top_10_Gaming_Laptop['Graphics Processor']=Graphics_Processor
Top_10_Gaming_Laptop['Seller']=Seller
Top_10_Gaming_Laptop['Price']=Price
Top_10_Gaming_Laptop
```
